## Задача:
#### Установить имеет ли компания в радиусе необходимого нам количества км хотя бы одну заправку нашей сети.

1. Импортируем таблицу с координатами заправок и таблицу с координатами клиентов. 

2. Далее зададим радиус( ввод самостоятельно ). 

3. Выведем точки на карту, где синими пинами будет обозначены заправки сети, а красными пинами будут обозначены клиенты.

4. Напишем функцию, в условии которой будет поиск заправки в заданом нами радиусе от клиента.

5. Выводим список компаний, которые находятся в заданном нами радиусе от заправки.

```python
import pandas as pd
import numpy as np
import folium
from geopy.distance import geodesic
```

```python
#Ссылка на данные с координатами заправок с локального пк


coordinates = pd.read_excel(r'C:\Users\dimak\OneDrive\Рабочий стол\аналитика\coordinates.xlsx')

#Ссылка на данные с координатами клиента с локального пк


clients = pd.read_excel(r'C:\Users\dimak\OneDrive\Рабочий стол\аналитика\clients_coordinates.xlsx')
```

```python
radius = input("Пожалуйста, введите радиус в км, и нажмите ENTER: ")
```

```python
radius = pd.to_numeric(radius)
```

```python
m = folium.Map(location=[58, 68], zoom_start=4)
for _, row in coordinates.iterrows():
    folium.Marker([row['Широта'], row['Долгота']], icon=folium.Icon(color='blue')
    ).add_to(m)
    
for _, row in clients.iterrows():
    folium.Marker(
        location=[row['Широта'], row['Долгота']],
        icon=folium.Icon(color='red')
    ).add_to(m)
```

```python
m
```

```python
def is_within_radius(client_lat, client_lon, coordinates_df, radius):
    client_point = (client_lat, client_lon)
    for _, row in coordinates_df.iterrows():
        station_point = (row['Широта'], row['Долгота'])
        if geodesic(client_point, station_point).kilometers <= radius:
            return True
    return False

clients['В радиусе'] = clients.apply(lambda row: is_within_radius(row['Широта'], row['Долгота'], coordinates, radius), axis=1)
```

```python
print(clients[clients['В радиусе'] == True]['В радиусе'].count(), " компаний имеют хотя бы 1 заправку в радиусе ", radius, "км.")
```
